```python
import pandas as pd
import regex

from database.comments import Comments
from database.topics import Topics

import sys
sys.path.append('../functions')
import preprocessing_fncs as prep

cs = Comments(env='dev')
tp = Topics(env='dev')
```

```python
df = cs.read_all()
df = df.copy()
df = prep.format_df(df, borough_col_name='council')
```

```python
def keyword_match(df, column, keyword):
    pattern = rf'\b{regex.escape(keyword)}s?\b'
    matched_comments = df[df[column].str.contains(pattern, flags=regex.IGNORECASE, regex=True, na=False)]
    n = len(matched_comments)
    print(f"Number of comments mentioning '{keyword}': {n}")
    
    return matched_comments
```

```python
keywords = ['fuck', 'shit', 'hate', 'bastard', 'communist', 'drugs', 'chavs', 'benefits', 'crack', 'race', 'immigrant']
```

```python
for word in keywords:
    keyword_match(df, 'cleaned_comment_text', word)
```

```python
shit_comments = keyword_match(df, 'cleaned_comment_text', 'shit')
with pd.option_context('display.max_colwidth', None):
    print(shit_comments['cleaned_comment_text'])
```

```python
immigrant_comments = keyword_match(df, 'cleaned_comment_text', 'immigrant')
with pd.option_context('display.max_colwidth', None):
    print(immigrant_comments['cleaned_comment_text'])
```
